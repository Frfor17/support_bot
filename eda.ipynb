```python
# –Ø—á–µ–π–∫–∞ 1: –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –ø–µ—Ä–≤—ã–π –≤–∑–≥–ª—è–¥
import pandas as pd
import os
from pathlib import Path
```

```python

# –£–∫–∞–∂–∏ –ø—É—Ç—å –∫ —Å–≤–æ–µ–π –ø–∞–ø–∫–µ —Å –¥–∞–Ω–Ω—ã–º–∏
DATA_PATH = "./datasets"  # <-- –ò–ó–ú–ï–ù–ò –ù–ê –°–í–û–ô –ü–£–¢–¨!

# –ü–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ —Ç–∞–º –µ—Å—Ç—å
files = list(Path(DATA_PATH).rglob("*.*"))
print(f"üìÅ –í—Å–µ–≥–æ —Ñ–∞–π–ª–æ–≤: {len(files)}")

for i, f in enumerate(files[:10]):  # –ü–æ–∫–∞–∂–µ–º –ø–µ—Ä–≤—ã–µ 10
    size_mb = f.stat().st_size / (1024*1024)
    print(f"{i+1}. {f.name} ({size_mb:.2f} MB, {f.suffix})")
```
